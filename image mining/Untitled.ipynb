```python
#Importing OpenCV - the computer vision library
import cv2
```

```python
path='/Users/StevenTseng/Desktop/img1.jpg'
```

```python
test=cv2.imread(str(path))
```

```python
test.shape
```

```python
gray=cv2.cvtColor(test,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray)
```

```python
sns.distplot(gray.flatten())
plt
```

```python
from skimage.filters import threshold_otsu
```

```python
thres=threshold_otsu(gray)
thres
```

```python
mask=np.where(gray>thres,1,0)
```

```python
plt.figure(figsize=(12,6))
plt.imshow(mask,cmap='gray')
plt.title('Edge Image')
plt.xticks([])
plt.yticks([])
```
